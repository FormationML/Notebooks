### Exercice 1 Tracez une courbe de y1 et y2 en fct de x : 
    - La courbe y1 doit être de couleur rouge
    - L'axe x doit s'apeller Temps 
    - L'axe y doit s'appeler Amplitude
    - Ajoutez les légentes "sinus" et "cosinus"

```python
x= np.linspace(0, 10, 100)
y1 = np.cos(x) 
y2 = np.sin(x)
```

```python
plt.plot(x, y1, color= "red",label="cosinus")
plt.plot(x ,y2,label="sinus")
plt.xlabel("Temps")
plt.ylabel("Amplitude")
plt.legend()
```

### Exercice 5 : Colormesh 

Tracez un spectrogramme

```python
# generate 2d grid for the x & y bounds
dx, dy = 0.05, 0.05
y, x = np.mgrid[slice(1, 5 + dy, dy),
                slice(1, 5 + dx, dx)]
z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)
```

```python
plt.pcolormesh(x, y, z, cmap='jet')
```

### Exercice 6 plot en 3D 

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(1, 12, 100)
y = x[:, np.newaxis]
im = y * np.sin(x) * np.cos(y)
```

 Visualisez l'image en 3D et en intéractif

```python
%matplotlib notebook
ax = plt.axes(projection='3d')
xgrid, ygrid = np.meshgrid(x, y.ravel()) # ravel -> returns contiguous flattened array 
ax.plot_surface(xgrid, ygrid, im, cmap=plt.cm.jet) # cmap -> palette de couleurs
```
